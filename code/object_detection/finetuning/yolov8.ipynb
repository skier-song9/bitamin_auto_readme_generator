## Load a pretrained model

```python
from ultralytics import YOLO
model = YOLO('./yolov8s.pt')
print(type(model.names), len(model.names))
print(model.names)
```

## Train a model

```python
model.train(data='/content/powerpoint-detector-2/data.yaml',epochs=200,patience=100, batch=64, optimizer = 'auto',lr0=0.01, dropout=0.1, imgsz=640, project = '/content/drive/MyDrive/README')
```
