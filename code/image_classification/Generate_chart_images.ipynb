```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Define plot path
plot_path = '../../data/image_classification/input/generated_charts'

assert os.path.exists(plot_path)

# Define plot types
plot_types = ['confusion_matrix_plots', 'heatmap_plots', 'box_plots', 'violin_plots']

# color palettes
cmaps = ['Blues','viridis','Reds','GnBu','PuBuGn','YlOrRd','tab10','Accent','rocket','crest','inferno']
len_cmaps = len(cmaps)

# general xlabel and ylabel
x_labels = [
    "Time", "Date", "Duration", "Elapsed Time", "Hour", "Minute", "Second",
    "Month", "Year", "Age", "Distance", "Frequency", "Temperature", "Pressure",
    "Depth", "Position", "Index", "Iteration", "Sample", "Step", "Phase",
    "Wavelength", "Longitude", "Latitude", "Period", "Cycle", "Session",
    "Trial", "Epoch", "Round"
]
y_labels = [
    "Value", "Measurement", "Amplitude", "Magnitude", "Intensity", "Count",
    "Frequency", "Rate", "Speed", "Velocity", "Acceleration", "Force",
    "Power", "Energy", "Mass", "Weight", "Height", "Length", "Width", "Depth",
    "Concentration", "Density", "Volume", "Pressure", "Temperature",
    "Humidity", "Angle", "Probability", "Percentage", "Score"
]
gen_colnames = ["id", "name", "date", "timestamp", "user_id", "username", "email", 
    "age", "gender", "location", "address", "phone_number", "birthdate", 
    "created_at", "updated_at", "status", "category", "type", "description", 
    "amount", "price", "quantity", "total", "score", "rating", "comments", 
    "feedback", "latitude", "longitude", "ip_address"]
len_labels = len(x_labels)

# Repeat code for each plot type
for plot_type in tqdm(plot_types):
    
    # Repeat code for 200 times
    for i in range(250):
        # generate random index of x,ylabels
        xlabel,ylabel = np.random.randint(0,len_labels), np.random.randint(0,len_labels)
        xlabel,ylabel = x_labels[xlabel], y_labels[ylabel]

        # select random color map
        cmap_index = np.random.randint(0,len_cmaps)
        cmap_ = cmaps[cmap_index]
        
        # Generate random data based on plot type
        if plot_type == 'confusion_matrix_plots':       
            col_num = np.random.randint(2,8)
            data = np.random.randint(0, 1000, size=(col_num,col_num))
            plt.figure(figsize=(8, 6))
            sns.heatmap(data, annot=True, fmt='d', cmap=cmap_)
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.title(f'Confusion Matrix {i+1}')
            
        elif plot_type == 'heatmap_plots':
            col_num = np.random.randint(6,20)
            data = np.random.rand(col_num,col_num)
            plt.figure(figsize=(10, 8))
            # randomly add annot parameter
            sns.heatmap(data, cmap=cmap_, annot=True ,fmt='.1f')
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.title(f'Heatmap {i+1}')
            
        elif plot_type == 'box_plots':
            xlabel,ylabel = np.random.randint(0,len_labels), np.random.randint(0,len_labels)
            xlabel,ylabel = x_labels[xlabel], y_labels[ylabel]
            col_num = np.random.randint(3,10)
            col_names = gen_colnames[:col_num]
            degrees_of_freedom = np.random.randint(3,10)
            means = np.random.uniform(-10, 10, col_num)  # 각 column의 평균
            std_devs = np.random.uniform(0.5, 5, col_num)  # 각 column의 표준편차
            data = np.array([np.random.normal(loc=means[i], scale=std_devs[i], size=1000) for i in range(col_num)]).T
            plt.figure(figsize=(8, 6))
            ax = sns.boxplot(data=data, palette=cmap_)
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            ax.set_xticklabels(col_names)
            plt.title(f'Box Plot {i+1}')
        
        elif plot_type == 'violin_plots':
            xlabel,ylabel = np.random.randint(0,len_labels), np.random.randint(0,len_labels)
            xlabel,ylabel = x_labels[xlabel], y_labels[ylabel]
            col_num = np.random.randint(3,10)
            degrees_of_freedom = np.random.randint(3,10)
            means = np.random.uniform(-10, 10, col_num)  # 각 column의 평균
            std_devs = np.random.uniform(0.5, 5, col_num)  # 각 column의 표준편차
            data = np.array([np.random.normal(loc=means[i], scale=std_devs[i], size=1000) for i in range(col_num)]).T
            plt.figure(figsize=(8, 6))
            sns.violinplot(data=data, cmap=cmap_)
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.title(f'Violin Plot {i+1}')
        
        # Save plot in the appropriate directory
        plt.savefig(os.path.join(plot_path, f'{plot_type[:-6]}_{i+1}.png'))
        plt.close()

print("200 plots for each type have been generated and saved.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Define plot path
plot_path = '../../data/image_classification/input/generated_charts'
assert os.path.exists(plot_path)

# Define plot types
plot_types = ['confusion_matrix_plots', 'heatmap_plots']

# color palettes
cmaps = ['Blues','viridis','Reds','GnBu','PuBuGn','YlOrRd','tab10','Accent','rocket','crest','inferno']
len_cmaps = len(cmaps)

# general xlabel and ylabel
x_labels = [
    "Time", "Date", "Duration", "Elapsed Time", "Hour", "Minute", "Second",
    "Month", "Year", "Age", "Distance", "Frequency", "Temperature", "Pressure",
    "Depth", "Position", "Index", "Iteration", "Sample", "Step", "Phase",
    "Wavelength", "Longitude", "Latitude", "Period", "Cycle", "Session",
    "Trial", "Epoch", "Round"
]
y_labels = [
    "Value", "Measurement", "Amplitude", "Magnitude", "Intensity", "Count",
    "Frequency", "Rate", "Speed", "Velocity", "Acceleration", "Force",
    "Power", "Energy", "Mass", "Weight", "Height", "Length", "Width", "Depth",
    "Concentration", "Density", "Volume", "Pressure", "Temperature",
    "Humidity", "Angle", "Probability", "Percentage", "Score"
]
gen_colnames = ["id", "name", "date", "timestamp", "user_id", "username", "email", 
    "age", "gender", "location", "address", "phone_number", "birthdate", 
    "created_at", "updated_at", "status", "category", "type", "description", 
    "amount", "price", "quantity", "total", "score", "rating", "comments", 
    "feedback", "latitude", "longitude", "ip_address"]
len_labels = len(x_labels)

# Repeat code for each plot type
for plot_type in plot_types:
    
    # Repeat code for 100 times
    for i in tqdm(range(100)):
        # generate random index of x,ylabels
        xlabel, ylabel = np.random.randint(0, len_labels), np.random.randint(0, len_labels)
        xlabel, ylabel = x_labels[xlabel], y_labels[ylabel]

        # select random color map
        cmap_index = np.random.randint(0, len_cmaps)
        cmap_ = cmaps[cmap_index]
        
        # Generate random data based on plot type
        if plot_type == 'confusion_matrix_plots':
            col_num = np.random.randint(2, 8)
            data = np.random.randint(0, 350, size=(col_num, col_num))
            
            # Increase diagonal values
            np.fill_diagonal(data, np.random.randint(700, 1000, size=col_num))
            
            plt.figure(figsize=(8, 6))
            sns.heatmap(data, annot=True, fmt='d', cmap=cmap_)
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.title(f'Confusion Matrix {i+1}')
            
        elif plot_type == 'heatmap_plots':
            col_num = np.random.randint(6, 20)
            # data = np.random.rand(col_num, col_num)
            data = np.random.normal(loc=0, scale=0.1, size=(col_num, col_num))
            
            # Set diagonal values to 1
            np.fill_diagonal(data, 1)
            
            plt.figure(figsize=(10, 8))
            # randomly add annot parameter
            sns.heatmap(data, cmap=cmap_, annot=True, fmt='.1f')
            plt.xlabel(xlabel)
            plt.ylabel(ylabel)
            plt.title(f'Heatmap {i+1}')
        
        # Save plot in the appropriate directory
        plt.savefig(os.path.join(plot_path, f'2_{plot_type[:-6]}_{i+1}.png'))
        plt.close()

print("200 plots for each type have been generated and saved.")
```
